# TASK #3: VISUALIZE DATASET

```python
employes_db["Attrition"]=employes_db["Attrition"].apply(lambda x:1 if x=="Yes" else 0)
```

```python
employes_db["Department"].value_counts()
```

```python
employes_db["Department"].replace("Research & Development",0,inplace=True)
employes_db["Department"].replace("Sales",1,inplace=True)
employes_db["Department"].replace("Human Resources",2,inplace=True)
employes_db["MaritalStatus"].replace("Married",1,inplace=True)
employes_db["MaritalStatus"].replace("Single",0,inplace=True)
employes_db["MaritalStatus"].replace("Divorced",2,inplace=True)
employes_db["Gender"]=employes_db["Gender"].apply(lambda x:1 if x=="Male" else 0)
employes_db["Over18"]=employes_db["Over18"].apply(lambda x:1 if x=="Y" else 0)
employes_db["OverTime"]=employes_db["OverTime"].apply(lambda x:1 if x=="Yes" else 0)
```

```python
employes_db["JobRole"].replace("Sales Executive",0,inplace=True)
employes_db["JobRole"].replace("Research Scientist",1,inplace=True)
employes_db["JobRole"].replace("Laboratory Technician",2,inplace=True)
employes_db["JobRole"].replace("Manufacturing Director",3,inplace=True)
employes_db["JobRole"].replace("Healthcare Representative",4,inplace=True)
employes_db["JobRole"].replace("Manager",5,inplace=True)
employes_db["JobRole"].replace("Sales Representative",6,inplace=True)
employes_db["JobRole"].replace("Research Director",7,inplace=True)
employes_db["JobRole"].replace("Human Resources",8,inplace=True)
```

```python
employes_db["EducationField"].replace("Life Sciences",0,inplace=True)
employes_db["EducationField"].replace("Medical",1,inplace=True)
employes_db["EducationField"].replace("Marketing",2,inplace=True)
employes_db["EducationField"].replace("Technical Degree",3,inplace=True)
employes_db["EducationField"].replace("Other",4,inplace=True)
employes_db["EducationField"].replace("Human Resources",5,inplace=True)
```

```python
employes_db["BusinessTravel"].replace("Travel_Rarely",0,inplace=True)
employes_db["BusinessTravel"].replace("Travel_Frequently",1,inplace=True)
employes_db["BusinessTravel"].replace("Non-Travel",2,inplace=True)
```

# TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
# TASK #4: CREATE TESTING AND TRAINING DATASET & PERFORM DATA CLEANING
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.3, random_state=101)
```

# TRAIN AND EVALUATE A DEEP LEARNING MODEL

```python
import tensorflow as tf 
```

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=500,activation="relu",input_shape=(30,)))
model.add(tf.keras.layers.Dense(units=500,activation="relu"))
model.add(tf.keras.layers.Dense(units=500,activation="relu"))
model.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
# oversampler = SMOTE(random_state=0)
# smote_train, smote_target = oversampler.fit_sample(X_train, y_train)
# epochs_hist = model.fit(smote_train, smote_target, epochs = 100, batch_size = 50)
epochs_hist=model.fit(X_train, Y_train,epochs=100,batch_size=50)
```

```python
Y_pred=model.predict(X_test)
Y_pred=(Y_pred>0.5)
Y_pred
```

```python
plt.plot(epochs_hist.history["loss"])
```

```python
plt.plot(epochs_hist.history["accuracy"])
```

```python
cm=confusion_matrix(Y_test,Y_pred)
sns.heatmap(cm,annot=True)
print(classification_report(Y_pred,Y_test))
```
